```python
import zipfile

# Updated path to include the subfolder
zip_path = "/content/food safety analyzer/FOOD ML.zip"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall("/content/")

print("Unzipped successfully!")
```

```python
import pandas as pd

df = pd.read_csv("/content/food_adulteration_data.csv")
df.head()
```

```python
df.info()
df['health_risk'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder

df = df.drop(columns=["adulteration_id"])

le = LabelEncoder()

for col in df.columns:
    df[col] = le.fit_transform(df[col])

df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

X = df.drop("health_risk", axis=1)
y = df["health_risk"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train, y_train)

pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, pred))
```

```python
import pickle

pickle.dump(model, open("model.pkl", "wb"))
```

```python
sample = X_test.iloc[0].values.reshape(1, -1)

prediction = model.predict(sample)

print("Predicted Risk Level:", prediction[0])
```

```python
risk_labels = ["Low", "Medium", "High"]

print("Predicted:", risk_labels[prediction[0]])
```
